```python
import pandas as pd

df = pd.read_csv('../dengs_fish/DengsFish.csv')
df = df.drop(df.columns[0], axis=1)
df.head()
```

```python
df_sentence = df.groupby('orth', sort=False).first().reset_index()[['orth', 't1_orth', 't2_orth']]
df_sentence.head()
```

```python
from pydub import AudioSegment

df_full = AudioSegment.from_wav('../dengs_fish/05OCT2013_OttoGwadoAyoker_DengsFish.wav')

for idx, row in df_sentence.iterrows():
    # Get the start and end time from the DataFrame (convert to milliseconds)
    start_time = row['t1_orth'] * 1000  # convert to milliseconds
    end_time = row['t2_orth'] * 1000    # convert to milliseconds
    
    # Extract the audio chunk
    audio_chunk = df_full[start_time:end_time]
    
    # Create a filename based on the row index (e.g., "chunk_0.wav")
    output_filename = f"df{idx}.wav"
    
    # Export the audio chunk to a file
    audio_chunk.export(output_filename, format="wav")

    print(f"Saved: {output_filename}")
```
